```python
import time
from colorama import Fore, Style
from PIL import Image
import pygame
import os

# Initialize pygame
pygame.init()

# Load sounds
sound_file_path = r"F:\Voluntary works\Code Workshop_OU\Submitted_material_Codeworkshop_2023\Course material\Day 2\dragon-roar-high-intensity-36564.mp3"
try:
    pygame.mixer.init()
    pygame.mixer.music.load(sound_file_path)
except pygame.error:
    print("Failed to load the sound file. Please check the file path.")

def load_image(image_filename):
    image_path = os.path.join("F:/Voluntary works/Code Workshop_OU/Submitted_material_Codeworkshop_2023/Course material/Day 2", image_filename)
    return Image.open(image_path)

def display_image(image):
    image.show()

def introduction(princess_image):
    print("Welcome to the Interactive Text Adventure Game!")
    time.sleep(1)
    print("You are a brave hero on a quest to rescue the kidnapped princess.")
    time.sleep(1)
    display_image(princess_image)  # Display the princess image
    time.sleep(1)
    print("You find yourself standing in front of a castle guarded by a fearsome dragon.")
    time.sleep(1)

def make_choice(prompt, choices):
    print(prompt)
    time.sleep(1)
    while True:
        for i, choice in enumerate(choices, start=1):
            print(f"{i}. {choice}")
        try:
            choice_number = int(input("Enter the number of your choice: "))
            if 1 <= choice_number <= len(choices):
                return choice_number - 1
            else:
                print("Invalid choice. Please enter a valid number.")
        except ValueError:
            print("Invalid input. Please enter a number.")

def dragon_encounter():
    choices = ["Fight the dragon", "Try to sneak past the dragon", "Run away"]
    choice = make_choice("Oh no! A fierce dragon appears! What do you want to do?", choices)

    if choice == 0:
        print("You engage in a fierce battle with the dragon.")
        time.sleep(1)
        print("After a tough fight, you manage to defeat the dragon and rescue the princess!")
        time.sleep(1)
        print(Fore.RED + "As a reward for your bravery, the kingdom presents you with a crown!")
        print(Style.RESET_ALL)
        print(Fore.RED + red_crown_reward)
        print(Style.RESET_ALL)
        # Play victory sound
        pygame.mixer.music.play()
    elif choice == 1:
        print("You try to sneak past the dragon, but it spots you and attacks!")
        time.sleep(1)
        print("You put up a good fight, but the dragon is too powerful. You lose.")
    elif choice == 2:
        print("You run away from the castle, leaving the princess behind.")
        time.sleep(1)
        print("You failed to rescue the princess. Game over.")

def show_medal_reward():
    print(Fore.CYAN + "Congratulations! You have received a medal for your bravery!")
    print(Style.RESET_ALL)
    print(Fore.CYAN + medal_reward)
    print(Style.RESET_ALL)

dragon_reward = r"""
           / \__
     (    @\___
      /         O
   /   (_____/
 /_____/   U
"""

red_crown_reward = r"""
      .~~.
     '.\ '.
     : .  :
     :   :
     :   :
    ,'   '.
  .'       '.
 :_________:
 :         :
 :         :
 :         :
 :_________:


"""

medal_reward = r"""
    _____
, -'     ` -,
 \         /  `
  ` -., - '
"""

def continue_adventure(dragon_image):
    choices = ["Explore the castle", "Enter the castle through the main gate"]
    choice = make_choice("You arrive at the castle entrance. What do you want to do?", choices)

    if choice == 0:
        print("You cautiously explore the surroundings and find a hidden passage.")
        time.sleep(1)
        print("You sneak inside the castle through the secret passage.")
        # Display the reward - medal
        show_medal_reward()
    elif choice == 1:
        print("You boldly march towards the main gate.")
        time.sleep(1)
        print("The dragon spots you and prepares to attack.")
        display_image(dragon_image)  # Display the dragon image
        pygame.mixer.music.play()  # Play the dragon sound

    dragon_encounter()

def main():
    # Load images
    princess_image = load_image("princess.png")  # Replace "princess.png" with the filename of your princess image file
    dragon_image = load_image("dragon.jpg")  # Replace "dragon.jpg" with the filename of your dragon image file

    introduction(princess_image)  # Pass the princess_image to the introduction function
    continue_adventure(dragon_image)  # Pass the dragon_image as an argument

if __name__ == "__main__":
    main()
```
